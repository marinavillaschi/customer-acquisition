```python
# setting random state for reproducibility
random_state = 22

# import libraries
import pandas as pd
```

The datasets provided for this projects are too large. 

To avoid problems as we develop the project, I've decided to take a sample of both datasets to work with.

These are the two data files that will be handled in this notebook:

- `Udacity_AZDIAS_052018.csv`: Demographics data for the general population of Germany; 891 211 persons (rows) x 366 features (columns).
- `Udacity_CUSTOMERS_052018.csv`: Demographics data for customers of a mail-order company; 191 652 persons (rows) x 369 features (columns).

```python
# load in the data
azdias_full = pd.read_csv('data/Udacity_AZDIAS_052018.csv', sep=';')
customers_full = pd.read_csv('data/Udacity_CUSTOMERS_052018.csv', sep=';')
```

```python
# check the datasets sizes
print('AZDIAS:     ',azdias_full.shape)
print('CUSTOMERS:  ',customers_full.shape)
```

```python
azdias_full.head()
```

```python
customers_full.head()
```

Let's go ahead and do a random sampling of 25% of the datasets with no replacing.

```python
azdias = azdias_full.sample(frac=0.25, replace=False, random_state=random_state)
customers = customers_full.sample(frac=0.25, replace=False, random_state=random_state)

print(f'Rows of full AZDIAS dataset:      {azdias_full.shape[0]}\nRows of sample AZDIAS dataset:    {azdias.shape[0]}\n')
print(f'Rows of full CUSTOMERS dataset:   {customers_full.shape[0]}\nRows of sample CUSTOMERS dataset: {customers.shape[0]}')
```

We should now have these two files to go ahead with the project:

- `sample_AZDIAS.csv`: Demographics data for the general population of Germany; 222 805 persons (rows) x 366 features (columns).
- `sample_CUSTOMERS.csv`: Demographics data for customers of a mail-order company; 47 913 persons (rows) x 369 features (columns).

```python
# exporting sample files to csv
azdias.to_csv('data/sample_AZDIAS.csv')
customers.to_csv('data/sample_CUSTOMERS.csv')

print('\n----> Sample datasets were exported successfully!')
```
